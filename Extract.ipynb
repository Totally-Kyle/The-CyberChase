```python
import opendatasets as od
import pandas as pd
import os

# Step 1: Download the dataset
dataset_url = 'https://www.kaggle.com/datasets/sibichakkaravarthys/vikrant-honeypot-data'
od.download(dataset_url)

# Step 2: Find actual data directory
root_dir = 'vikrant-honeypot-data'
all_files = []

# Walk through all subdirectories to get all CSV paths
for dirpath, _, filenames in os.walk(root_dir):
    for file in filenames:
        if file.endswith('.csv'):
            all_files.append(os.path.join(dirpath, file))

# Step 3: Target date strings
target_dates = ['2023-02-05', '2023-02-06', '2023-02-07', '2023-02-08',
                '2023-02-09', '2023-02-10', '2023-02-11', '2023-02-12']

# Step 4: Match files to dates and load into variables
for date_str in target_dates:
    matched = [f for f in all_files if date_str in os.path.basename(f)]
    if matched:
        try:
            df = pd.read_csv(matched[0])
            var_name = f"df_{date_str.replace('-', '_')}"
            globals()[var_name] = df
            print(f"Loaded {matched[0]} into `{var_name}` — shape: {df.shape}")
        except Exception as e:
            print(f"Error reading {matched[0]}: {e}")
    else:
        print(f"No file found for date {date_str}")
```
